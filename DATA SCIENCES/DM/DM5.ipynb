```python
import pandas as pd
from datetime import datetime
import os
```

```python
# Récupération données 
path = 'https://www.data.gouv.fr/fr/datasets/r/'
liste = ['38adef39-797c-4ec0-924a-09f053a6f246','9e68e978-9cfe-4eaf-ab9c-95196f382d3f','f7642f6d-809e-421c-893c-10c8c52f7340','89ba3cdb-8628-44c1-91f4-9c535fa1b09c','88c2f643-2ece-4f70-a38d-cb032b2a21b9','2406184c-7880-4172-bb49-c14e04845207','0b281be1-b81a-4d62-82b8-84f0ec43271e','fd90ca39-2a30-4751-b0aa-cdfb2f4433cc']
data = pd.DataFrame()

for element in liste:
    data = data.append(pd.read_csv(path+element, sep=";",encoding = "ISO-8859-1"))
    
data.head()
```

```python
data[data['dep_mon'] != '0,00']
```

```python
data['dep_mon'] = data['dep_mon'].str.replace('.','').str.replace(',','.')
data['rec_mon'] = data['rec_mon'].str.replace('.','').str.replace(',','.')
data['rem_mon'] = data['rem_mon'].str.replace('.','').str.replace(',','.')
```

```python
data['dep_mon'] = data['dep_mon'].astype(float)
data['rec_mon'] = data['rec_mon'].astype(float)
data['rem_mon'] = data['rem_mon'].astype(float)
```

```python
data[data['dep_mon'] > 0]
```

```python
data['depassement'] = data['dep_mon'].apply(lambda x: True if x > 0 else False)
```

```python
pv = pd.pivot_table(data[['l_pre_spe','depassement','rem_date']], index=['l_pre_spe','depassement'], aggfunc='count').reset_index()
pv
```

```python
pv_depassement = pv.pivot(index = 'l_pre_spe', columns = 'depassement', values = 'rem_date')
pv_depassement
```
