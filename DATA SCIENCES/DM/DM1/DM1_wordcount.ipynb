```python


import sys
import os

# +++your code here+++
# Define print_words(filename) and print_top(filename) functions.
# You could write a helper utility function that reads a file
# and builds and returns a word/count dict for it.
# Then print_words() and print_top() can just call the utility function.

def count_words(file_handle):
  """Count word frequencies in a text file"""
  frequencies = {}
  for line in file_handle.readlines():
    for word in line.split():
      word = word.casefold()
      if word in frequencies:
        frequencies[word] += 1
      else:
        frequencies[word] = 1
  return frequencies

def print_words(filename):
  """ Print word occurrences in a text file"""
  file_handle = open(filename, 'r')
  if file_handle.mode == 'r':
    frequencies = count_words(file_handle)
    for word in sorted(frequencies):
      print(word, frequencies[word])
  return

def print_top(filename):
  """ Print the 20 words that occur more in text file"""
  file_handle = open(filename, 'r')
  if file_handle.mode == 'r':
    frequencies = count_words(file_handle)
    for word in sorted(frequencies, reverse=True, key=frequencies.__getitem__)[:20]:
      print(word, frequencies[word])
  return

###

# This basic command line argument parsing code is provided and
# calls the print_words() and print_top() functions which you must define.
def main():
  if len(sys.argv) != 3:
    print('usage: ./wordcount.py {--count | --topcount} file')
    sys.exit(1)
  print(os.getcwd())

  option = sys.argv[1]
  filename = sys.argv[2]
  if option == '--count':
    print_words(filename)
  elif option == '--topcount':
    print_top(filename)
  else:
    print('unknown option: ' + option)
    sys.exit(1)

if __name__ == '__main__':
  main()
```
